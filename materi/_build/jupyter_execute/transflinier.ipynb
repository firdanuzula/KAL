# Transformasi Linier

## Buktikan bahwa matriks A adalah transformasi linier

Transformasi:

$$T(x) = Ax$$

dengan

$$A = \begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}$$
dan

\
$$x = \begin{bmatrix}
x \\
y
\end{bmatrix}$$

Hitung Operasi Matrik:

\
$$T(x) = 
\begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta 
\end{bmatrix}$ 
$\begin{bmatrix}
x \\
y \end{bmatrix} $
=
$\begin{bmatrix}
x \cos \theta - y \sin \theta \\
x \sin \theta + y \cos \theta 
\end{bmatrix}$$

Sifat 1:

\
$$T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})
\$$

Misal:
\
$$\mathbf{u} = \begin{bmatrix} x_1 \\ y_1 \end{bmatrix}, \quad \mathbf{v} = \begin{bmatrix} x_2 \\ y_2 \end{bmatrix}$$

Hitung:
\
$$T(\mathbf{u} + \mathbf{v}) = T\left( \begin{bmatrix} x_1 + x_2 \\ y_1 + y_2 \end{bmatrix} \right)
= $\begin{bmatrix}
(x_1 + x_2)\cos \theta - (y_1 + y_2)\sin \theta \\
(x_1 + x_2)\sin \theta + (y_1 + y_2)\cos \theta
\end{bmatrix}$$

Sementara itu:
\
$T(\mathbf{u}) + T(\mathbf{v})$ =
$\begin{bmatrix}
x_1 \cos \theta - y_1 \sin \theta \\
x_1 \sin \theta + y_1 \cos \theta
\end{bmatrix}$
+
$\begin{bmatrix}
x_2 \cos \theta - y_2 \sin \theta \\
x_2 \sin \theta + y_2 \cos \theta
\end{bmatrix}$
=
$\begin{bmatrix}
(x_1 + x_2)\cos \theta - (y_1 + y_2)\sin \theta \\
(x_1 + x_2)\sin \theta + (y_1 + y_2)\cos \theta
\end{bmatrix}$


✅ Hasilnya sama $\rightarrow$ sifat 1 terpenuhi.

Sifat 2:

Misal:
\
cu = $\begin{bmatrix} cx_1 \\ cy_1 \end{bmatrix}$

Hitung:
\
T(cu) = $\begin{bmatrix} cx_1 \cos \theta - cy_1 \sin \theta \\ cx_1 \sin \theta + cy_1 \cos \theta \end{bmatrix}$$
= c\begin{bmatrix} x_1 \cos \theta - y_1 \sin \theta \\ x_1 \sin \theta + y_1 \cos \theta \end{bmatrix}$
= cT(u)

✅ Hasilnya sama $\rightarrow$ sifat 2 terpenuhi.

Karena kedua sifat tersebut terbukti berlaku untuk transformasi ini, maka:
\
$T(\mathbf{x}) = A\mathbf{x}$
dengan
\
$$A = \begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}$$

adalah transformasi linier di $\mathbb{R}^2$.
